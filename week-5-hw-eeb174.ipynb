```python
#Question 1
import csv
import sys

my_dict={}

#to read a csv file
file_object=open("formattednature.csv", encoding = "ISO-8859-1")
reader = csv.reader(file_object)
#remove headers,next goes to the next line
next(reader, None)
next(reader, None)
next(reader, None)

#for every row we take the first and third column which is the species and then families names. 
for row in reader:
    
    my_dict[row[0]]=row[2]
         
for keys, values in my_dict.items():
         print(keys+" belongs to Family "+values)
     


file_object.close()
```

```python
#Question 3
#first used this command in shell to clean up the file, since each , separates columns:
# sed 's/,\s/ /g' formattedebird.csv > ebird2.csv

from collections import Counter
import csv
import sys

file_object=open("ebird2.csv", encoding = "ISO-8859-1")

reader = csv.reader(file_object)

mylist=[]

next(reader, None)
next(reader, None)
next(reader, None)


#counter will make a dictionary of the families and the number of them repeated. 
for row in reader:
    
    mylist.append(row[7])
    
counted = Counter(mylist)    

for keys, values in counted.items():
        print(str(values)+" total species in Family "+ keys)
     


file_object.close()


```

```python
#Question 1
def protein_amino (protein,amino):
    protein= protein.upper()
    amino= amino.upper()
    protein_length= len(protein)
    amino_length= protein.count(amino)
    pa_content = amino_length *100/ protein_length
    return pa_content
#first we make both protein and amino in upper case to avoid mistakes. then we find the length of protein and number of aminos in it and then amino content.
```

```python
assert protein_amino("MSRSLLLRFLLFLLLLPPLP", "M")==5
```

```python
assert protein_amino("MSRSLLLRFLLFLLLLPPLP", "M") == 5
assert protein_amino("MSRSLLLRFLLFLLLLPPLP", "r") == 10
assert protein_amino("msrslllrfllfllllpplp", "L") == 50
assert protein_amino("MSRSLLLRFLLFLLLLPPLP", "Y") == 0
```

```python
#Question 2
#to get amino as a list we define it in function as a given list, if the second argument is not given then the function will use the amino list defined below. 
def protein_amino2 (protein,amino= ["A", "I", "L", "M", "F", "W", "Y", "V"]):
    protein= protein.upper()
    protein_length= len(protein)
    count = 0
    for letter in amino:
        letter= letter.upper()
        amino_length= protein.count(letter)
        count = amino_length +count
    pa_content = count *100 / protein_length
    return pa_content
```

```python
assert protein_amino2("MSRSLLLRFLLFLLLLPPLP", ["M"]) == 5
assert protein_amino2("MSRSLLLRFLLFLLLLPPLP", ['M', 'L']) == 55
assert protein_amino2("MSRSLLLRFLLFLLLLPPLP", ['F', 'S', 'L']) == 70
assert protein_amino2("MSRSLLLRFLLFLLLLPPLP") == 65
```
